```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("datasetnamin.csv")
data.head()
```

```python
plt.hist(data['Have you encountered technical difficulties or challenges while participating in online classes, such as internet connectivity issues or platform-related problems?'], bins=20, edgecolor='k')
plt.xlabel('Response')
plt.ylabel('Frequency')
plt.title('Have you encountered technical difficulties or challenges while participating in online classes, such as internet connectivity issues or platform-related problems?')
plt.show()
```

```python
plt.hist(data['Do you believe that face-to-face classes provide a better opportunity for interactive discussions and engagement with peers compared to online classes?'], bins=20, edgecolor='k')
plt.xlabel('Response')
plt.ylabel('Frequency')
plt.title('Do you believe that face-to-face classes provide a better opportunity for interactive discussions and engagement with peers compared to online classes?')
plt.show()
```
